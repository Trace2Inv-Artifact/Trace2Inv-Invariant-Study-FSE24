```python
import matplotlib.pyplot as plt
import json
%matplotlib inline
```

```python
with open("createSchain.json") as f:
    data = json.load(f)
    
x = [measurement['nodesAmount'] for measurement in data]
y = [measurement['gasUsed'] for measurement in data]
plt.figure(figsize=(15, 10))
plt.ticklabel_format(useOffset=False, style='plain')
plt.title("Create schain gas usage")
plt.xlabel("Amount of nodes")
plt.ylabel("Gas used")
plt.ylim(0, 12e6)
plt.plot(x, y)
plt.show()
```
